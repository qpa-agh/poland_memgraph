```python
import geopandas as gpd
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from shapely import wkt
```

```python
df_trees = pd.read_csv('ads24-trees.csv', low_memory=False)
df_trees
```

```python
crs = {'init': 'epsg:4326'}
df_trees['geometry'] = df_trees['wkt'].apply(wkt.loads)

gdf_trees = gpd.GeoDataFrame(df_trees, crs=crs, geometry='geometry')
gdf_trees['long'] = gdf_trees.geometry.x
gdf_trees['lat'] = gdf_trees.geometry.y
gdf_trees
```

```python
gdf_trees.to_crs(epsg=2180, inplace=True)
gdf_trees['x'] = gdf_trees.geometry.x
gdf_trees['y'] = gdf_trees.geometry.y
gdf_trees
```

```python
gdf_trees = gdf_trees.drop(['geometry', 'wkt'], axis=1)
```
